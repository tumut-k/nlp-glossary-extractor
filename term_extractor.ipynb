```python
# Version 2.0: Lemmatization and POS Tagging added for better linguistic accuracy
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download required NLTK datasets (wordnet is the lexical database for lemmatization)
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')

# Define the source text
source_text = "Translation is not just about words... Translation, as we know it, is about meaning! A good translator knows the meaning (and context) 100%."

# 1. Tokenization
tokenized_words = word_tokenize(source_text.lower())

# 2. Part-of-Speech (POS) Tagging: Identifies if a word is a noun, verb, adjective, etc.
# Example output format: [('words', 'NNS'), ('knows', 'VBZ')]
tagged_words = nltk.pos_tag(tokenized_words)

stop_words = stopwords.words('english')
lemmatizer = WordNetLemmatizer()

# Helper function to translate NLTK's POS tags into WordNet's POS tags for accurate lemmatization
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN # Default fallback is noun

term_frequencies = {}

# 3. Filter and Lemmatize
for word, tag in tagged_words:
    # Remove punctuation and stop words
    if word.isalpha() and word not in stop_words:

        # Get the grammatical category of the word (Verb, Noun, etc.)
        wn_tag = get_wordnet_pos(tag)

        # Reduce the word to its dictionary root (Lemma) based on its POS tag
        lemma = lemmatizer.lemmatize(word, pos=wn_tag)

        # Count the frequency of the root words
        if lemma in term_frequencies:
            term_frequencies[lemma] += 1
        else:
            term_frequencies[lemma] = 1

# 4. Output the V2.0 terminology frequencies
print("Version 2.0 - Lemmatized Terminology Extraction Results:")
print(term_frequencies)
```
